```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range of x from -5 to 5
x = np.linspace(-5 * np.pi, 5 * np.pi, 1000)

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(x, x, label='x' , color = 'b')
plt.plot(x, -x, label='-x', color ='b')
plt.plot(x, np.sin(x), label='sin(x)', color ='g', linewidth=2)
plt.plot(x, x * np.sin(x), label='x*sin(x)', color ='r', linewidth=3)

# Add legend
plt.legend(loc ='upper right')

# Annotate the origin
plt.annotate('origin', xy=(0, 0))

# Set axes limits
plt.xlim(-5 * np.pi, 5 * np.pi)
plt.ylim(min(x * np.sin(x)) - 1, max(x * np.sin(x)) + 1)

# Add grid and labels
plt.grid(False)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Summary')

# Show the plot
plt.show()
```
