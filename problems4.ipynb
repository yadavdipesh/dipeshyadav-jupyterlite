3. Look up the documentation for `plt.errorbar` (you can look this up online
   also) and draw the means of each batsman along with an error bar
   illustrating the standard deviation of their scores.

```python
import numpy as np
import matplotlib.pyplot as plt

lst_category = ["dhoni","gambhir","kohli","rohit","sachin","sehwag","kapil dev",]

arr_dhoni = np.loadtxt('C:/Users/dyadav/ePGDhomeexercise/data/dhoni.txt')
arr_gambhir = np.loadtxt('C:/Users/dyadav/ePGDhomeexercise/data/gambhir.txt')
arr_kohli = np.loadtxt('C:/Users/dyadav/ePGDhomeexercise/data/kohli.txt')
arr_rohit = np.loadtxt('C:/Users/dyadav/ePGDhomeexercise/data/rohit.txt')
arr_sachin = np.loadtxt('C:/Users/dyadav/ePGDhomeexercise/data/sachin.txt')
arr_sehwag = np.loadtxt('C:/Users/dyadav/ePGDhomeexercise/data/sehwag.txt')
arr_kapil_dev = np.loadtxt('C:/Users/dyadav/ePGDhomeexercise/data/kapil_dev.txt')

lst_mean = []

#mean
lst_mean.append(np.mean(arr_dhoni))
lst_mean.append(np.mean(arr_gambhir))
lst_mean.append(np.mean(arr_kohli))
lst_mean.append(np.mean(arr_rohit))
lst_mean.append(np.mean(arr_sachin))
lst_mean.append(np.mean(arr_sehwag))
lst_mean.append(np.mean(arr_kapil_dev))

#Std daviation
lst_std = []
lst_std.append(np.std(arr_dhoni))
lst_std.append(np.std(arr_gambhir))
lst_std.append(np.std(arr_kohli))
lst_std.append(np.std(arr_rohit))
lst_std.append(np.std(arr_sachin))
lst_std.append(np.std(arr_sehwag))
lst_std.append(np.std(arr_kapil_dev))

#lst_std, lst_mean
yerr= 0.1
plt.errorbar(lst_mean,lst_std, yerr = yerr, ecolor= "red", capsize=5)
plt.title("means of each batsman along with an error bar illustrating the standard deviation of their scores")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

lst_category = ["dhoni","gambhir","kohli","rohit","sachin","sehwag","kapil dev",]

arr_dhoni = np.loadtxt('C:/Users/dyadav/ePGDhomeexercise/data/dhoni.txt')
arr_gambhir = np.loadtxt('C:/Users/dyadav/ePGDhomeexercise/data/gambhir.txt')
arr_kohli = np.loadtxt('C:/Users/dyadav/ePGDhomeexercise/data/kohli.txt')
arr_rohit = np.loadtxt('C:/Users/dyadav/ePGDhomeexercise/data/rohit.txt')
arr_sachin = np.loadtxt('C:/Users/dyadav/ePGDhomeexercise/data/sachin.txt')
arr_sehwag = np.loadtxt('C:/Users/dyadav/ePGDhomeexercise/data/sehwag.txt')
arr_kapil_dev = np.loadtxt('C:/Users/dyadav/ePGDhomeexercise/data/kapil_dev.txt')

# --- Plot 1: Dhoni vs gambhir ---
plt.figure(figsize=(10,4))
plt.hist([arr_dhoni, arr_gambhir], bins=8, label=['Dhoni','gambhir'], color=['blue','orange'], alpha=0.7)
plt.xlabel("Runs")
plt.ylabel("Frequency")
plt.title("Dhoni vs gambhir Scores")
plt.legend()
plt.show()

# --- Plot 2: Kohli vs rohit ---
plt.figure(figsize=(10,4))
plt.hist([arr_kohli, arr_rohit], bins=8, label=['kohli','rohit'], color=['green','red'], alpha=0.7)
plt.xlabel("Runs")
plt.ylabel("Frequency")
plt.title("Kohli vs rohit Scores")
plt.legend()
plt.show()

# --- Plot 3: Sachin vs Sehwag Vs Kapil dev ---
plt.figure(figsize=(10,4))
plt.hist([arr_sachin, arr_sehwag, arr_kapil_dev], bins=10, label=['Sachin','Sehwag', 'Kapil Dev'], color=['yellow','red', 'purple'], alpha=0.7)
plt.xlabel("Runs")
plt.ylabel("Frequency")
plt.title("Sachin vs Sehwag Vs Kapil dev Scores")
plt.legend()
plt.show()
```
